```python
import pandas as pd
```

```python
action_movies = pd.read_csv('action.csv')
action_movies
```

```python
action_movies.rename(columns={'gross(in $)': 'gross'}, inplace=True)
```

```python
action_movies.drop(columns=['star_id', 'director_id', 'description', 'movie_id', 'star', 'gross'], inplace=True)
```

```python
action_movies
```

```python
action_movies.rating.isna
```

```python
action_movies['rating'].fillna('unknown', inplace=True)
```

```python
action_movies.rating.unique
```

```python
action_movies = action_movies[action_movies['rating'] != 'unknown']
```

```python
action_movies
```

```python
action_movies['certificate'].fillna('unknown', inplace=True)
```

```python
action_movies
```

```python
print(action_movies['votes'].dtype)
```

```python
action_movies['votes'] = action_movies['votes'].astype(int)
```

```python
action_movies
```

```python
# Suponiendo que action_movies es tu DataFrame y las columnas se llaman 'votes', 'movie_name' y 'rating'
top_movies = action_movies[['movie_name', 'votes', 'rating']].sort_values(by='votes', ascending=False).head(10)
print(top_movies)
```

```python
# Ordenamos las pel√≠culas por la cantidad de votos en orden descendente
sorted_movies = action_movies.sort_values(by='votes', ascending=False)

# Seleccionamos la pel√≠cula con la mayor calificaci√≥n de entre las primeras
top_rated_movie = sorted_movies.iloc[0]

# Mostramos la pel√≠cula con la mayor calificaci√≥n entre las m√°s votadas
print("Pel√≠cula con m√°s votos y mayor rating:")
print("Nombre:", top_rated_movie['movie_name'])
print("Votos:", top_rated_movie['votes'])
print("Rating:", top_rated_movie['rating'])
```

```python
# Ordenamos las pel√≠culas por la cantidad de votos en orden descendente
sorted_movies = action_movies.sort_values(by='votes', ascending=False)

# Seleccionamos las 10 pel√≠culas con m√°s votos
top_10_movies = sorted_movies.head(10)

# De las 10 pel√≠culas seleccionadas, encontramos aquella con el mayor rating
top_rated_movie = top_10_movies.loc[top_10_movies['rating'].idxmax()]

# Mostramos las 10 pel√≠culas con m√°s votos y la pel√≠cula con el mayor rating entre ellas
print("Las 10 pel√≠culas con m√°s votos:")
print(top_10_movies[['movie_name', 'votes', 'rating']])
print("\nPel√≠cula con m√°s votos y mayor rating:")
print("Nombre:", top_rated_movie['movie_name'])
print("Votos:", top_rated_movie['votes'])
print("Rating:", top_rated_movie['rating'])
```
